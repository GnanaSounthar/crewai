```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from  crewai import LLM
llm= LLM(
    model="gemini/gemini-2.0-flash",
    temperature=0.1
)
llm.call("who is virat kohli")
```

```python
from crewai import Agent,Task,Crew
email_assitant=Agent(
    role="email assitant agent",
    goal="Improve emails and make them sound professional and clear",
    backstory="A highly experienced communication expert skilled in professional email writing",
    verbose=True,
    llm=llm
)

original_email=("hey team, just want to tell projects so kindly submit the task previous task ,ready to the project ,we get the specifi roll to the project, "
                "let me know this project ,thankyou guys ")

email_task =Task(
              description=f"""Take the following rough email and rewrite it into a professional and polished
     version.
           Expand abbreviation:
           '''{original_email}'''""",
              agent=email_assitant,
              expected_output="A professional  written email with proper formatting and content"
          )
crew =Crew(
         agents=[email_assitant],
         tasks=[email_task],
         verbose=True
     )

result =crew.kickoff()
print(result)

```
